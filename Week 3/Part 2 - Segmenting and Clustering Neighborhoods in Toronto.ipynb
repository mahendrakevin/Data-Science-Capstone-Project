Import and read Toronto dataframe

```python
df_toronto = pd.read_csv('df_toronto.csv')
df_toronto.head()
```

Import and read geospatial data

```python
df_coords = pd.read_csv('Geospatial_data.csv')
df_coords.head()
```

Add Latitude & Longitude column to Toronto dataframe and fill the value from geospatial data

```python
df_toronto['Latitude'] = np.nan
df_toronto['Longitude'] = np.nan

for idx in df_toronto.index:
    coords_idx = df_coords['Postal Code'] == df_toronto.loc[idx, 'PostalCode']
    df_toronto.at[idx, 'Latitude'] = df_coords.loc[coords_idx, 'Latitude'].values
    df_toronto.at[idx, 'Longitude'] = df_coords.loc[coords_idx, 'Longitude'].values

df_toronto.head(20)
```

Export Toronto Dataframe to CSV file

```python
df_toronto.to_csv('df_toronto2.csv', index=False)
```
